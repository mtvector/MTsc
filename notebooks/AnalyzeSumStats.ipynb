```python
import pandas as pd
import os
import seaborn
import matplotlib.pyplot as plt
```

```python
def p2f(x):
    if(type(x)==str):
        return float(x.strip('%'))/100
    else:
        return x
```

# First Sequencing

```python
pathto='/home/mt/Downloads/macaque2/'
```

```python
dfs=[]
samplenames=[]

for d in os.listdir(pathto):
    dfs.append(pd.read_csv(pathto+d+'/outs/metrics_summary.csv',thousands=','))
    samplenames.append(d)
```

```python
df=pd.concat(dfs,axis=0,sort=False)
df.index=samplenames
```

```python
df.fillna(0,inplace=True)
```

```python
percs=[x for x in df.columns if type(df.loc[:,x][0])==str]

for i in percs:
    df.loc[:,i]=[p2f(x) for x in df.loc[:,i]]
```

```python
df.shape
```

```python
df=df.loc[sorted(df.index),:]
```

```python
df
```

```python
#df.to_csv(pathto+"sumstats.csv")
```

```python
import plotly
import plotly.plotly as py
import plotly.graph_objs as go

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
for x in df.columns:
    data = [go.Bar(x=df.index,
            y=df[x])]
    layout = go.Layout(
    title=go.layout.Title(
        text=x,
        xref='paper',
        x=0
    ))
    fig = go.Figure(data=data, layout=layout)
    iplot(fig, filename=x, show_link=False)
```

```python
data = [go.Scatter(x=df['Sequencing Saturation'],
        y=df['Median UMI Counts per Cell'],mode= 'markers',marker=dict(size= 14,
                    line= dict(width=1),
                    opacity= 0.3
                   ),name= x,
        text= df.index)]
layout = go.Layout(
title=go.layout.Title(
    text='Saturation vs Median UMI',
    xref='paper',
    x=0
),    xaxis= dict(
        title= 'Saturation',
        ticklen= 5,
        zeroline= False,
        gridwidth= 2,
    ),
    yaxis=dict(
        title= 'Median UMI Per Cell',
        ticklen= 5,
        gridwidth= 2,
    ),
    showlegend= False)
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename=x, show_link=False)
```
