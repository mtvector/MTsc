```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
#(Reads-umis)/reads
def calcSaturation(L1):
    return((L1.loc[:,'NUM_GENIC_READS']-L1.loc[:,'NUM_TRANSCRIPTS'])/L1.loc[:,'NUM_GENIC_READS'])
```

```python
L1= pd.read_csv('/home/mt/Downloads/Chimp/summary/Chimp_S2_L001_dge.summary.txt',sep='\t',skiprows=4,header=1)
L2=pd.read_csv('/home/mt/Downloads/Chimp/summary/Chimp_S2_L002_dge.summary.txt',sep='\t',skiprows=4,header=1)
```

```python
L1= L1.loc[L1.loc[:,'NUM_GENIC_READS']>0,:]
L2= L2.loc[L2.loc[:,'NUM_GENIC_READS']>0,:]
```

```python
#plt.hist(L1.loc[:,'NUM_GENIC_READS']/L1.loc[:,'NUM_TRANSCRIPTS'])
plt.hist(calcSaturation(L1))
print(np.mean(calcSaturation(L1)))
```

```python
plt.plot(np.cumsum(L1.loc[:,'NUM_GENIC_READS'])/np.sum(L1.loc[:,'NUM_GENIC_READS']))
plt.xlim((-1, 300))
plt.title('Cumulative density of reads Chimp')
```

```python
plt.plot(np.cumsum(L1.loc[:,'NUM_GENIC_READS'])/np.sum(L1.loc[:,'NUM_GENIC_READS']))
plt.title('Cumulative density of reads Chimp')
```
