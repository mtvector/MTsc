```python
import pandas as pd
import os
import seaborn
import matplotlib.pyplot as plt
```

```python
pathto='/home/mt/Downloads/macaque2/'
```

```python
dfs=[]
samplenames=[]
def p2f(x):
    return float(x.strip('%'))/100

for d in os.listdir(pathto):
    dfs.append(pd.read_csv(pathto+d+'/outs/metrics_summary.csv',thousands=',',converters={'col':p2f}))
    samplenames.append(d)
```

```python
df=pd.concat(dfs,axis=0)
df.index=samplenames
```

```python
df
```

```python
for i in df.columns[4:-2]:
    df.loc[:,i]=[p2f(x) for x in df.loc[:,i]]
```

```python
df.shape
```

```python
df=df.loc[sorted(df.index),:]
```

```python
sorted(df.columns)
```

```python
for x in df.columns:
    seaborn.barplot(df.index,df[x]).set_title(x)
    plt.show()
```
